## Description

This mechanism resembles a Stephenson III type mechanism, with a difference in the 5-link loop, as a revolute joint is replaced with a prismatic one, resulting in the slider, located with $r_6$. (Pending reference to Stephenson applications).

## Variables and parameters

main variable: $\theta_2$

secondary variable: $\theta_3,\theta_4,\theta_5,r_6,$

parameters: $r_2,r_3,r_{4a},r_{4b},r_{5},r_{1a},r_{1b},\beta$

## Vector ecuations

$\overrightarrow{r_2} + \overrightarrow{r_{4a}} - \overrightarrow{r_3} - \overrightarrow{r_{1a}} = 0$ (1)

$\overrightarrow{r_3} + \overrightarrow{r_{4b}} + \overrightarrow{r_5} - \overrightarrow{r_{6}} - \overrightarrow{r_{1b}}= 0$ (2)

## Scalar equations
$r2*cos(\theta_2)+r4a*cos(\theta_4)-r3*cos(\theta_3)-r1a=0$ (3)

$r2*sin(\theta_2)+r4a*sin(\theta_4)-r3*sin(\theta_3)=0$ (4)

$r3*cos(\theta_3)+r4b*cos(\theta_4+\beta)+r5*cos(\theta_4)-r1b=0$ (5)

$r3*sin(\theta_3)+r4b*sin(\theta_4+\beta)+r5*sin(\theta_5)-r6=0$ (6)

## Code implementation

### Import modulues

```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

### Function definition

```python
def six_bar(initCond,params):
    # parameters and main variable
    th2 = params[0]
    r2 = params[1]
    r3 = params[2]
    r4a = params[3]
    r4b = params[4]
    r5 = params[5]
    r1a = params[6]
    r1b = params[7]
    beta = params[8]
    # initial conditions
    th3 = initCond[0]
    th4 = initCond[1]
    th5 = initCond[2]
    r6 = initCond[3]
    
    Y = np.zeros(4)
    Y[0] = r2*np.cos(th2) + r4a*np.cos(th4) - r3*np.cos(th3) - r1a;
    Y[1] = r2*np.sin(th2) + r4a*np.sin(th4) - r3*np.sin(th3);
    Y[2] = r3*np.cos(th3) + r4b*np.cos(th4+np.pi-beta) + r5*np.cos(th5) - r1b;
    Y[3] = r3*np.sin(th3) + r4b*np.sin(th4+np.pi-beta) + r5*np.sin(th5) - r6;
    return Y
```

### Function test

```python
th2 = np.deg2rad(60)
r2=13
r3=14
r4a=11
r4b=7
r5=15
r1a=20
r1b=7
beta=np.deg2rad(62.96)

parameters = [th2,r2,r3,r4a,r4b,r5,r1a,r1b,beta]

# initial condition
th3Ini=np.deg2rad(101.5)
th4Ini=np.deg2rad(12.9)
th5Ini=np.deg2rad(17.9)
r6Ini=23.688
x0 = [th3Ini,th4Ini,th5Ini,r6Ini]

print(six_bar(x0,parameters))
```

## Solve for initial condition

### Using Fsolve to analize the initial conditions

```python
q,info,exitflag,message = fsolve(six_bar,x0,args = parameters,full_output = True)

print("solution2 : [Rad(th3 | th4 | th5) / mm (r6)]= ",q,"\n",exitflag,"\n",message)
th3 = q[0]
th4 = q[1]
th5 = q[2]
r6y = q[3]
```

### Point definition for mechanism graphic representation

```python
# Point Definition
Ox = 0
Oy = 0
Ax = r2*np.cos(th2)
Ay = r2*np.sin(th2)
Cx = Ax + r4a*np.cos(th4)
Cy = Ay + r4a*np.sin(th4)
Bx = Cx - r3*np.cos(th3)
By = Cy - r3*np.sin(th3)
Dx = Cx + r4b*np.cos(th4+np.pi-beta)
Dy = Cy + r4b*np.sin(th4+np.pi-beta)
Ex = Dx + r5*np.cos(th5)
Ey = Dy + r5*np.sin(th5)
```

```python
# graph of the mechanism 
fig = plt.figure(1)

#r2
hr2 = plt.plot([Ox,Ax],[Oy,Ay])
#r3
hr3 = plt.plot([Cx,Bx],[Cy,By])
#r4
hr4a = plt.plot([Ax,Cx],[Ay,Cy])
hr4b = plt.plot([Cx,Dx],[Cy,Dy])
hr4c = plt.plot([Ax,Dx],[Ay,Dy])
#r5
hr5 = plt.plot([Dx,Ex],[Dy,Ey])

plt.axis("scaled")
plt.xlim(-r2,r2+r3+r5)
plt.ylim(-r2,r2+r4b+r5)
plt.grid()
plt.draw() 
```

## Solution for a range of values

We use a cyclic programming structure that allows us to execute the above several times. Next we define how many values we want to evaluate.

```python
numpos = 100
```

### Solution arrays definition

```python
posInicial = np.deg2rad(26)
posFinal = np.deg2rad(67)
th2V = np.linspace(posInicial,posFinal,numpos)
```

### Solution for the defined range

```python
th3V = np.zeros(numpos)
th4V = np.zeros(numpos)
th5V = np.zeros(numpos)
r6yV = np.zeros(numpos)
```

```python
exitflagV = np.zeros(numpos)
```

```python
for i in range(0,numpos):
    parameters[0] = th2V[i]
    
    # resolvemos con fsolve
    q,info,exitflagV[i],mensaje = fsolve(six_bar,x0,args = parameters,full_output = True)
    
    th3V[i] = q[0]
    th4V[i] = q[1]
    th5V[i] = q[2]
    r6yV[i] = q[3]
    x0 = q
    
    if exitflagV[i] != 1:
        print('animation error ' + str(i))
        print(message)
 
```

## Animation

```python
# Point Definition
POx = 0
POy = 0
AxV = r2*np.cos(th2V)
AyV = r2*np.sin(th2V)
CxV = AxV + r4a*np.cos(th4V)
CyV = AyV + r4a*np.sin(th4V)
BxV = CxV - r3*np.cos(th3V)
ByV = CyV - r3*np.sin(th3V)
DxV = CxV + r4b*np.cos(th4V+np.pi-beta)
DyV = CyV + r4b*np.sin(th4V+np.pi-beta)
ExV = DxV + r5*np.cos(th5V)
EyV = DyV + r5*np.sin(th5V)
```

### Preparing the animation
To make animations we will use the animation module of matplotlib.

```python
from matplotlib.animation import FuncAnimation
```

```python
def animate(i):
    hr2xi= np.array([Ox,AxV[i]])
    hr2yi = np.array([Oy,AyV[i]])
    
    hr3xi = np.array([CxV[i],Bx])
    hr3yi = np.array([CyV[i],By])
    
    hr4axi = np.array([AxV[i],CxV[i]])
    hr4ayi = np.array([AyV[i],CyV[i]])
    hr4bxi = np.array([CxV[i],DxV[i]])
    hr4byi = np.array([CyV[i],DyV[i]])
    hr4cxi = np.array([AxV[i],DxV[i]])
    hr4cyi = np.array([AyV[i],DyV[i]])
    
    hr5xi = np.array([DxV[i],Ex])
    hr5yi = np.array([DyV[i],EyV[i]])
    
    hr2[0].set_xdata(hr2xi)
    hr2[0].set_ydata(hr2yi)
    hr3[0].set_xdata(hr3xi)
    hr3[0].set_ydata(hr3yi)
    hr4a[0].set_xdata(hr4axi)
    hr4a[0].set_ydata(hr4ayi)
    hr4b[0].set_xdata(hr4bxi)
    hr4b[0].set_ydata(hr4byi)
    hr4c[0].set_xdata(hr4cxi)
    hr4c[0].set_ydata(hr4cyi)
    hr5[0].set_xdata(hr5xi)
    hr5[0].set_ydata(hr5yi)
```

```python
duration = 5000
animation = FuncAnimation(fig, animate, interval=duration/numpos, save_count=numpos)
```
