---

- Nota sobre el uso de *fsolve*

Para usar fsolve debemos **definir una función** que reciba los siguientes argumentos:

1. una lista de **valores (condiciones) iniciales** de las variables secundarias (incógnitas),

![funcion generica](funcion_generica.svg)

que en código se vería así:

```python
def nombre_funcion(x0):
    valInicial1,valInicial2,..., valInicialN = x0
    return [
        ecuacion1,
        ecuacion2,
        ...,
        ecuacionN
    ]
```

donde ecuacion1, ecuacion2,..., ecuacionN son expresiones que corresponden a las ecuaciones escalares del sistema no lineal que queremos resolver.

**Nota 1:** al llamar a esta función con valores de parámetros y variables correspondientes a una posición del mecanismo (es decir, una solución verdadera), debe obtener una lista de valores cercanos a cero.

<span class="mark">**Nota 2:** la lista de condiciones iniciales **debe ser ingresada en el mismo orden que se utiliza dentro de la función**</span>

---

### Definir función

#### Argumentos de la función (entradas)

x0 = [$\theta_{3ini}$,$r_{4ini}$] : **lista de numpy (array)** con estimados de las variables secundarias para la posición de la barra de entrada especificada

#### Otros datos que requiere la función

$\theta_2,r2, r3, \theta_{4}$ : **variables** previamente definidas en celdas anteriores, con los valores correspondientes

#### Salidas de la función

**lista de numpy (array)** con la evaluación de las ecuaciones escalares. 

Si las condiciones iniciales son adecuadas, el resultado debe ser una lista con valores cercanos a cero.

La función para este mecanismo se vería así:

![funcion manivela corredera](manivela_corredera.svg)

**Atención**: El número de condiciones iniciales (elementos de la lista x0) debe ser IGUAL al número de ecuaciones escalares

````python
def manivela_corredera(x0):
    '''
    Esta es la ayuda de la función.
    Condiciones iniciales: th3 y r4
    '''
    # Condiciones iniciales - esto se hace para mejorar legibilidad
    th3,r4 = x0
    #Versión legible de las ecuaciones
    return [
        r2*np.cos(th2) - r3*np.cos(th3) - r4*np.cos(th4),
        r2*np.sin(th2) - r3*np.sin(th3) - r4*np.sin(th4)
    ]
````

````python
help(manivela_corredera)
````

### Definir parámetros

Los parámetros se definen como variables, con el nombre que usted elija para cada una.

**CUIDADO: Mantenga los nombres de los parámetros usados en la función.**

````python
th2 = np.deg2rad(0)
r2 = 25
r3 = 87.5
th4 = 0

print("theta2 = ", th2)
````

### Definir CONDICIONES INICIALES

Son valores aproximados de $\theta_3$ y $r_4$ para la posición de la barra de entrada especificada.

Se pueden obtener usando método gráfico o por inspección del mecanismo real.

Entre más se aproximen a la realidad mejor.

**CUIDADO: mantenga el órden utilizado en la función. En este caso, th3 va de primero, luego r4**

````python
th3 = np.pi
r4 = 112.5
x0 = [th3,r4]
#la lista de condiciones iniciales debe ser 
#ingresada en el mismo orden que se utiliza dentro de la función
````

### Probar condiciones iniciales en la función

Debemos obtener valores cercanos a CERO

````python
manivela_corredera(x0)
````

### Probar la función usando fsolve()

llamar a 
```python
q,info,exitFlag,mensaje = fsolve(-nombre de la función-,
                                 -condiciones iniciales-,
                                 full_output = True)
```

q = lista con la solución (valores de las incógnitas $\theta_3$ y $r_4$)

info = diccionario con información del proceso

exitFlag = número entero, 1 si converge, diferente si no converge

mensaje = cadena de texto con información sobre el resultado

````python
q,info,exitFlag,mensaje = fsolve(manivela_corredera,
                                x0,
                                full_output = True)

print(mensaje,"la respuesta es ",q)
print("las condiciones iniciales x0 son: ",x0)
````

Si las condiciones iniciales provistas fueron correctamente medidas, deben ser similares a la solución encontrada.

## Solución para un rango de valores

Objetivo: resolver el análisis cinemático en un rango de valores de la variable de entrada $\theta_2$, tal que

$\theta_2 = [0°:360°]$ con numpos = 100 posiciones

para esto podemos utilizar:

- linspace de numpy
- range de Python
- arange de numpy

nos será más útil la función linspace ya que podemos indicar el número de valores a analizar.

````python
numpos = 360  #número de posiciones a analizar
th2ini = th2  #posición inicial para el analisis. Debo poner la posición de la barra de entrada que definí al principio
th2fin = th2ini + 2*np.pi #posición final de la barra de entrada. En este caso le sumo 2pi para analizar una rev. completa
th2v = np.linspace(th2ini,th2fin,numpos)

print(np.rad2deg(th2v))
````

### iterar sobre la variable de entrada

primero creamos una lista vacia para cada variable secundaria. Allí almacenaremos las respuestas

luego, iteramos sobre la lista de la variable de entrada th2v

````python
th3v = []
r4v = []
# la función enumerate() entrega la posición y el valor de la lista que se está iterando.
for th2 in th2v: #esto actualiza el valor de la variable de entrada para cada iteración
    #los demás parámetros NO se actualizan.
    #resolvemos el sistema de ecuaciones usando fsolve. 
    #El resultado se almacena en variables temporales th3 y r4
    ### CUIDADO: el orden de las condiciones iniciales debe ser el mismo de la funcion
    [th3,r4],info,exitFlag,mensaje = fsolve(manivela_corredera,
                                    x0,
                                    full_output = True)
    #almacenar la respuesta en los arrays que creamos arriba
    #la función lista.append(valor) agrega un valor al final de la lista
    th3v.append(th3)
    r4v.append(r4)
    #actualizar las condiciones iniciales
    x0 = [th3,r4]
    if exitFlag != 1:
        print("Fallo en th2 = ",np.rad2deg(th2))
        print(mensaje,exitFlag)
        #si fallo, el numero de posiciones calculadas es menor que el numero de posiciones deseadas(numpos)
        #actualizo el numero de posiciones
        numpos = len(th3v)
        #luego, recortar el vector (lista) de variable de entrada para que quede del mismo tamaño
        th2v = th2v[0:numpos]
        break
np.rad2deg(th3v)
r4v
# print(len(th3v))
````

## graficar variables secundarias vs. variable de entrada

$\theta_2$ es la variable de entrada, por ende es la variable independiente (eje x)

$\theta_3$ es la variable secundaria, es decir, la variable dependiente (eje y)

````python
import matplotlib.pyplot as plt

fig = plt.figure()
plt.plot(np.rad2deg(th2v),np.rad2deg(th3v))
plt.grid()
plt.xlabel("$\\theta_2$ [°]")
plt.ylabel("$\\theta_3$ [°]")
fig2 = plt.figure()
plt.plot(np.rad2deg(th2v),r4v)
plt.grid()
plt.xlabel("$\\theta_2$ [°]")
plt.ylabel("$r_4$ [mm]")
````

## Animación

### Ubicación de puntos para animación



Para hacer la animación, **debemos encontrar la posición de todos los puntos que conforman el mecanismo ($pO,pA,PB$) PARA CADA POSICIÓN DE LA BARRA DE ENTRADA A ANALIZAR (numpos)** 

Primero calculamos TODAS las posiciones de los puntos A y B, utilizando las cantidades vectoriales ```
th2v, th3v, r4v
```

````python
Ox = 0
Oy = 0
Axv = r2*np.cos(th2v)
Ayv = r2*np.sin(th2v)
Bxv = Axv - r3*np.cos(th3v)
Byv = Ayv - r3*np.sin(th3v)

print("el tamaño de Axv es ",len(Axv))
````

## Posición inicial del mecanismo

debemos identificar cada elemento del gráfico con un nombre, en este caso

```python
hr2 = plt.plot(...)
```

hr2 es un identificador de la barra r2 (azul)

````python
posicion = 100   #es un número entre 0 y numpos-1
# crear la figura
fig = plt.figure()
# graficar las barras
hr2 = plt.plot([Ox,Axv[posicion]],
               [Oy,Ayv[posicion]])
hr3 = plt.plot([Bxv[posicion],Axv[posicion]],
               [Byv[posicion],Ayv[posicion]])
# organizar la figura
# escalar los ejes (misma escala)
plt.axis("scaled")
# definir limites
lim_inf_x = -r2
lim_sup_x = r2 + r3
lim_inf_y = -r2
lim_sup_y = r2
plt.xlim(lim_inf_x,lim_sup_x)
plt.ylim(lim_inf_y,lim_sup_y)
# cuadrícula
plt.grid()
````

## Animación

Debemos importar el módulo de animaciones 

````python
from matplotlib.animation import FuncAnimation
````

### Función de animación

Definimos la función que dibuja cada cuadro (plot)

````python
def animar(pos):
    #definir posición de puntos usando listas de numpy
    manivelaX = [Ox,Axv[pos]]
    manivelaY = [Oy,Ayv[pos]]
    acopladorX = [Bxv[pos],Axv[pos]]
    acopladorY = [Byv[pos],Ayv[pos]]
    #modificar la posición de las barras previamente creadas en la gráfica (fig)
    hr2[0].set_xdata(manivelaX)
    hr2[0].set_ydata(manivelaY)
    hr3[0].set_xdata(acopladorX)
    hr3[0].set_ydata(acopladorY)
    # Retornar siempre las barras que se están moviendo.
    return hr2[0],hr3[0]
````

### Configuración y ejecución de animación

Debemos especificar:

- longitud del video en milisegundos
- número máximo de cuadros (si especificamos muchos, el video quedará muy pesado)

Luego ejecutar la animación con la función FuncAnimation()

````python
videoLength = 5000 # en milisegundos
maxFrames = 80

increment = int(np.ceil(numpos/maxFrames))
totalFrames = range(0,numpos,increment)
animacion = FuncAnimation(fig, 
                          animar, 
                          frames = totalFrames,
                          interval=videoLength/len(totalFrames),
                          save_count=len(totalFrames),
                          blit=True)
````
