# Python Interactivo con Cuadernos de Jupyter

## Introducción

Esto es un cuaderno de Jupyter: un entorno interactivo de computación para crear documentos tipo cuaderno.

En un cuaderno puedes crear:

- Código ejecutable
- Widgets interactivos
- Gráficas (Plots)
- Texto narrativo
- Ecuaciones
- Imagenes
- Video

Estos documentos proveen un **registro autocontenido de computación** que puede ser convertido a varios formatos y compartido con otros por múltiples medios, incluso presentado libremente en la web (con GitHub o nbviewer.jupyter.org).
