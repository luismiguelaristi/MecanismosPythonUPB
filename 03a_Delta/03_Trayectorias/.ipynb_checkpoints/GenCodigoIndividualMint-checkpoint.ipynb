# Generador de código para Mint (robot delta)

Este cuaderno recibe un archivo individual .csv con los ángulos que el robot debe ejecutar
Genera un archivo "solutions.txt" que contiene el código que se debe ejecutar en el programa Mint del robot.

1. Abrir el archivo .csv con los ángulos

```python
%reset -sf
import tkinter as tk
from tkinter import filedialog

import os
from DeltaLib import genMintCode
import shutil

root = tk.Tk()
root.wm_attributes('-topmost', 1)
root.withdraw()

file_path = filedialog.askopenfilename()


if file_path != "":
    print(file_path)
else:
    print("Archivo no válido")
```

2. Revisar ángulos

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

with open(file_path) as f:
    reader = csv.reader(f)
    data = list(reader)

th = np.transpose(np.array(data))
# print(th)
th1v = np.array([float(x) for x in th[0][:]])
th2v = np.array([float(x) for x in th[1][:]])
th3v = np.array([float(x) for x in th[2][:]])
# print(th1v)
numpos = len(th1v)
# th1v
fig2,ax = plt.subplots(figsize=(15,5))
ax.plot([0,numpos],[-20,-20])
ax.plot([0,numpos],[-85,-85])
ax.set_title("Posición angular de los motores")
ax.set_xlabel("punto de trayectoria")
ax.set_ylabel("posición angular [°]")
ax.grid()
ax.plot(th1v,'.r')
ax.plot(th2v,'.g')
ax.plot(th3v,'.b')
ax.legend(["límite sup.","límite inf.","$\\theta_1$","$\\theta_2$","$\\theta_3$"])
```
