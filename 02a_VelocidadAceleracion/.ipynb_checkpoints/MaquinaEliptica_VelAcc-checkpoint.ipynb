## Kinematic analysis

### Mobility (Degrees of Freedom)

$n = 4$, $f_1 = 4$, $f_2 = 0$

\begin{equation}
\label{eq:1}
F = 3(n-1)-2f_1-f_2
\end{equation}

given $n$, $f_1$ and $f_2$, the DOF of the studied mechanism are obtained from evaluating (1)

$F = 1$ (2).

### Parameters and variables

Parameters: $r_2,r_3,r_4,r_{1x},r_{1y}$

Main variables: $\theta_2$

Secondary variables: $\theta_3, \theta_4$

### Vector diagram



### Vector equations

$\overrightarrow{r_2}-\overrightarrow{r_3}-\overrightarrow{r_4}-\overrightarrow{r_{1y}}+\overrightarrow{r_{1x}} = 0$

### Scalar equations

$r_2\cdot cos(\theta_{2})-r_3\cdot cos(\theta_{3})-r_4\cdot cos(\theta_{4})-r_{1y}\cdot cos(90°)+r_{1x}\cdot cos(0°)=0$,

$r_2\cdot sin(\theta_{2})-r_3\cdot sin(\theta_{3})-r_4\cdot sin(\theta_{4})-r_{1y}\cdot sin(90°)+r_{1x}\cdot sin(0°)=0$

which, after reducing terms, yields


$r_2\cdot cos(\theta_{2})-r_3\cdot cos(\theta_{3})-r_4\cdot cos(\theta_{4})+r_{1x}=0$,

$r_2\cdot sin(\theta_{2})-r_3\cdot sin(\theta_{3})-r_4\cdot sin(\theta_{4})-r_{1y}=0$

### Initial Conditions Measurement



Approximated values of $\theta_3$ and $\theta_4$ were measured for a specific position of link 2 (crank):

Variable|Value
:---:|:---:
$\theta_2$|**161.24°**
$\theta_3$|**10.38°**
$\theta_4$|76.95°+180° = **256.95°**

## función para implementar las ecuaciones escalares


los parámetros y la variable de entrada fueron definidos previamente en una celda más arriba

las condiciones iniciales son una lista (array de numpy)

definamos la lista de condiciones iniciales

````python
# mis_params = {th2= np.radians(161), # variable de entrada
#               r2= 3,  # parametros
#               r3= 12,
#               r4= 10,
#               th1x= np.radians(0),
#               th1y= np.radians(90),
#               r1x= 12,
#               r1y= 9}
th2= np.radians(161) # variable de entrada
r2= 3  # parametros
r3= 12
r4= 10
th1x= np.radians(0)
th1y= np.radians(90)
r1x= 12
r1y= 9
# variables secundarias
th3_ini = np.radians(10.5)
th4_ini = np.radians(257)
mi_x0 = [th3_ini,th4_ini]
````

````python
 # la función eliptica evalua las ecuaciones las condiciones iniciales dadas (x0)


def eliptica(x0):
    # OBTENER los valores de las incógnitas desde x0 es obligatorio, sino, fsolve no va a hacer nada
    th3 = x0[0]
    th4 = x0[1]
    return [r2*np.cos(th2) - r3*np.cos(th3) - r4*np.cos(th4) - r1y*np.cos(th1y) + r1x*np.cos(th1x),
            r2*np.sin(th2) - r3*np.sin(th3) - r4*np.sin(th4) - r1y*np.sin(th1y) + r1x*np.sin(th1x)]
    

eliptica(mi_x0)
````

## Solución con métodos numéricos

ahora vamos a usar fsolve

````python
from scipy.optimize import fsolve
````

````python
# help(fsolve)
th2= th2 + np.radians(30) 
sol,info,exitflag,mensaje = fsolve(eliptica,mi_x0,full_output=True)
# con diccionario de parámetros
# sol,info,exitflag,mensaje = fsolve(eliptica,mi_x0,args = params,full_output=True)
th3 = sol[0]
th4 = sol[1]
mi_x0 = sol

# verificamos que fsolve si haya resuelto
if exitflag != 1:
    print(exitflag)
    print(mensaje)
else:
    print(np.degrees(sol))
````

## Graficar una posición del mecanismo

vamos a posicionar las barras en el plano





````python
pOx = 0
pOy = 0
pAx = r2*np.cos(th2)
pAy = r2*np.sin(th2)
pBx = pAx - r3*np.cos(th3)
pBy = pAy - r3*np.sin(th3)
pCx = pBx - r4*np.cos(th4)
pCy = pBy - r4*np.sin(th4)
````

## Análisis para un rango de movimiento

Definamos un rango de variación de la variable de entrada, así como una cantidad de pasos para el análisis

````python
pos_ini = th2 # inicio en la posición que definí para th2 (el de la foto)
pos_fin = th2 + 2*np.pi
num_pos = 100
````

Ahora definimos un vector con esas características para la variable de entrada $\theta_2$

````python
print(np.degrees(pos_ini),np.degrees(pos_fin))
th2v=np.linspace(pos_ini,pos_fin,num_pos)
th2v
````

## Solución de las ecuaciones para dicho rango de movimiento

````python
# inicializar los vectores (listas) donde se van a guardar los resultados
th3v = []
th4v = []
for th2 in th2v:
    sol,info,exitflag,mensaje = fsolve(eliptica,mi_x0,full_output=True)
    # con diccionario de parámetros
    # sol,info,exitflag,mensaje = fsolve(eliptica,mi_x0,args = params,full_output=True)
    mi_x0 = sol
    # SI NO hubo una solución correcta
    if exitflag != 1:
        # PARE la ejecución y avise
        print(exitflag)
        print(mensaje)
        print("fallo en th2 = ",np.degrees(th2))
        break
    # SINO
    else:
        # Guarde los valores encontrados en las respectivas listas
        th3v.append(sol[0])
        th4v.append(sol[1])
    # Actualice condiciones iniciales
    
print(th3v)
num_pos = len(th3v)
th2v = th2v[0:num_pos]
# una vez terminado, actualice el número de posiciones que tuvieron solución:
# num_pos
# variable de entrada
th2v
````

````python
# len(th3v)
plt.plot(np.degrees(th2v),np.degrees(th3v))
plt.xlabel('$\\theta_2$ [°]')
plt.ylabel('$\\theta_3$ [°]')
plt.grid()
plt.title('Posición angular de la acopladora')
````

## Análisis de velocidad

1. definir derivada con respecto al tiempo de la variable de entrada

$\dot{\theta_2} = 2\pi$ rad/s  (yo defino este valor)

````python
th2p = 2*np.pi # manivela rota con velocidad constante
````

2. derivar las ecuaciones de posición con respecto al tiempo

$ r_{2} \cos{\left(\theta_{2} \right)} - r_{3} \cos{\left(\theta_{3} \right)} - r_{4} \cos{\left(\theta_{4} \right)} + r_{1x}$

$ \  r_{2} \sin{\left(\theta_{2} \right)} - r_{3} \sin{\left(\theta_{3} \right)} - r_{4} \sin{\left(\theta_{4} \right)} - r_{1y}$

ecuaciones derivadas:

$ - r_{2} \sin{\left(\theta_{2} \right)} \dot{\theta}_{2} + r_{3} \sin{\left(\theta_{3} \right)} \dot{\theta}_{3} + r_{4} \sin{\left(\theta_{4} \right)} \dot{\theta}_{4}$

$\  r_{2} \cos{\left(\theta_{2} \right)} \dot{\theta}_{2} - r_{3} \cos{\left(\theta_{3} \right)} \dot{\theta}_{3} - r_{4} \cos{\left(\theta_{4} \right)} \dot{\theta}_{4}$

3. implementar las ecuaciones en forma matricial ($Ax=b$)

$A=\left[\begin{matrix}+ r_{3} \sin{\left(\theta_{3} \right)} & r_{4} \sin{\left(\theta_{4} \right)}\\-r_{3} \cos{\left(\theta_{3} \right)} & - r_{4} \cos{\left(\theta_{4} \right)}\end{matrix}\right], b = \  \left[\begin{matrix}\dot{\theta_2} r_{2} \sin{\left(\theta_{2} \right)}\\- \dot{\theta_2} r_{2} \cos{\left(\theta_{2} \right)}\end{matrix}\right]$


### Calculemos velocidad para una posición específica del mecanismo

definimos la variable pos para indicar una posición del mecanismo.

vamos a utilizar una sola posición para este análisis

````python
pos = 25
th2 = th2v[pos]
th3 = th3v[pos]
th4 = th4v[pos]
pAx = pAx[pos]
pAy = pAy[pos]
pBx = pBx[pos]
pBy = pBy[pos]

b = np.array( [th2p*r2*np.sin(th2),
              -th2p*r2*np.cos(th2)] )
A = np.array([[r3*np.sin(th3),r4*np.sin(th4)],
              [-r3*np.cos(th3),-r4*np.cos(th4)]])
b,A
````

4. resolver con np.linalg.solve(A,b)

````python
x = np.linalg.solve(A,b)
th3p,th4p = x
th3p,th4p
````

en x tengo los valores de $\dot{\theta_3}$ y $\dot{\theta_4}$ respectivamente, para la posición de th2 dada

 ## Verificación numérica

 

````python
y2 = th4v[pos+1]
y1 = th4v[pos-1]
x2 = th2v[pos+1]
x1 = th2v[pos-1]
dth2 = x2-x1
dth4 = y2-y1

th4pnum = (dth4/dth2)*th2p
y2 = th3v[pos+1]
y1 = th3v[pos-1]
dth3 = y2-y1
th3pnum = (dth3/dth2)*th2p

th3pnum,th4pnum
````

 ### Velocidad de un punto
 
 

````python
w2 = np.array([0,0,th2p])
r2v = np.array([r2*np.cos(th2),r2*np.sin(th2),0])
velA = np.cross(w2,r2v)
````

Para calcular la velocidad del punto B, necesito w3 y r3



debemos invertir el vector r3 para poder usarlo con el modelo

````python
w3 = np.array([0,0,th3p])
# le sumo 180° a la dirección para invertir el vector.
r3_prima_v = np.array([r3*np.cos(th3+np.pi),r3*np.sin(th3+np.pi),0])
velRot3 = np.cross(w3,r3_prima_v)
velB = velA + velRot3
velB
````

````python

# Definimos figura para luego hacer la animación
figura = plt.figure(figsize=(10,6))
### manivela
barra_OA = plt.plot([pAx,pOx],[pAy,pOy])
#acopladora
barra_AB = plt.plot([pAx,pBx],[pAy,pBy])
#balancin
barra_BC = plt.plot([pCx,pBx],[pCy,pBy])
#vectores de velocidad
plt.quiver(pAx, pAy, velA[0], velA[1],scale_units='xy', scale=4)
plt.quiver(pBx, pBy, velA[0], velA[1],scale_units='xy', scale=4,color='b')
plt.quiver(pBx, pBy, velRot3[0], velRot3[1],scale_units='xy', scale=4,color='r')
plt.quiver(pBx, pBy, velB[0], velB[1],scale_units='xy', scale=4)

plt.axis('equal')
plt.xlim([-20,5])
plt.ylim([-10,15])
barra_AB
````

## Análisis de velocidad con Matemática Simbólica (Sympy)

Importar módulos para matemática simbólica.

La segunda línea es para que sympy use la notación de Newton para las variables dependientes del tiempo (punto)

````python
import sympy as sym
from sympy.physics.mechanics import dynamicsymbols, init_vprinting

sym.init_printing(use_unicode=False, wrap_line=True)
init_vprinting()
````

luego definimos:
- la variable independiente (el tiempo)
- las cantidades que NO dependen del tiempo se definen como símbolos simples
- las cantidades que SI dependen del tiempo, se definen como FUNCIONES, por eso va el argumento del final en th2s, th3s y th4s.

````python
t = sym.Symbol('t')
# Parámetros
r2s = sym.Symbol('r_2')
r3s = sym.Symbol('r_3')
r4s = sym.Symbol('r_4')
r1xs = sym.Symbol('r_{1x}')
r1ys = sym.Symbol('r_{1y}')
# Variables definidas como funcion para luego hacerlas depender de t
th2s, th3s, th4s = sym.symbols('theta_2 theta_3 theta_4', cls=sym.Function)

th2s(t),th3s(t),th4s(t)
````

Ahora derivamos con respecto al tiempo.

Cuidado con las variables dependientes del tiempo, deben ir siempre con paréntesis y la variable independiente por dentro.

````python
# Primera derivada de las variables con respecto al tiempo
th2sp = sym.diff(th2s(t),t)
th3sp = sym.diff(th3s(t),t)
th4sp = sym.diff(th4s(t),t)
th2spp = sym.diff(th2s(t),t,t)
th3spp = sym.diff(th3s(t),t,t)
th4spp = sym.diff(th4s(t),t,t)

th2sp,th3sp,th4sp,th2spp,th3spp,th4spp
````

Ahora definimos las ecuaciones escalares (las que están en la función) en términos de los parámetros y variables simbólicos.

```r2*np.cos(th2) - r3*np.cos(th3) - r4*np.cos(th4) - r1y*np.cos(th1y) + r1x*np.cos(th1x)```

```r2*np.sin(th2) - r3*np.sin(th3) - r4*np.sin(th4) - r1y*np.sin(th1y) + r1x*np.sin(th1x)]```

````python
eq1 = r2s*sym.cos(th2s(t)) - r3s*sym.cos(th3s(t)) - r4s*sym.cos(th4s(t)) + r1xs*sym.cos(sym.pi);
eq2 = r2s*sym.sin(th2s(t)) - r3s*sym.sin(th3s(t)) - r4s*sym.sin(th4s(t)) - r1ys;
eq1,eq2
````

ahora derivamos las ecuaciones

````python
eq1p = sym.diff(eq1,t)
eq2p = sym.diff(eq2,t)
eq1p,eq2p
````

Hasta aquí podemos comparar lo que derivamos a mano con este resultado.

Si querémos que Sympy nos resuelva el sistema, debemos hacer varias cosas primero.

### Definir los símbolos de las variables derivadas

````python
th2ssym, th3ssym, th4ssym = sym.symbols('theta_2 theta_3 theta_4')
th2spsym, th3spsym, th4spsym = sym.symbols('\dot{\\theta_2} \dot{\\theta_3} \dot{\\theta_4}')
th2sppsym, th3sppsym, th4sppsym = sym.symbols('\ddot{\\theta_2} \ddot{\\theta_3} \ddot{\\theta_4}')
th2ssym,th2spsym,th2sppsym
````

### Cambio de variables en las ecuaciones

esto se hace para cada ecuación del sistema

````python
eq1psym = eq1p.subs([
                  (th2sp,th2spsym),
                  (th3sp,th3spsym),
                  (th4sp,th4spsym)
                   ])
eq2psym = eq2p.subs([
                  (th2sp,th2spsym),
                  (th3sp,th3spsym),
                  (th4sp,th4spsym)
                    ])
eq1psym,eq2psym
````

### Conformación de matrices A y b

Con las ecuaciones bien conformadas, se puede llevar a la forma matricial

````python
eqns = [eq1psym,eq2psym]
As, bs = sym.linear_eq_to_matrix(eqns, [th3spsym,th4spsym])
As,bs
````

### Reemplazar valores escalares en las matrices

````python
As_eval = As.subs([(th3s(t),th3v[pos]),
                (th4s(t),th4v[pos]),
                  (r3s,r3),
                  (r4s,r4)])
bs_eval = bs.subs([(th2s(t),th2v[pos]),
              (th2spsym,th2p),
                  (r2s,r2)])
As_eval,bs_eval
````

### Solución para esos valores

Simplemente se resuelve con LUsolve de sympy

````python
[th3p_eval,th4p_eval] = As_eval.LUsolve(bs_eval)
(th3p_eval,th3p),(th4p_eval,th4p)
````

## Análisis de aceleración

Después de derivar por segunda vez las ecuaciones de posición con respecto al tiempo, y llevarlo a la forma matricial, obtenemos

$\displaystyle \left( \left[\begin{matrix}r_{3} \sin{\left(\theta_{3} \right)} & r_{4} \sin{\left(\theta_{4} \right)}\\- r_{3} \cos{\left(\theta_{3} \right)} & - r_{4} \cos{\left(\theta_{4} \right)}\end{matrix}\right], \  \left[\begin{matrix}\ddot{\theta_2} r_{2} \sin{\left(\theta_{2} \right)} + r_{2} \cos{\left(\theta_{2} \right)} \dot{\theta}_{2}^{2} - r_{3} \cos{\left(\theta_{3} \right)} \dot{\theta}_{3}^{2} - r_{4} \cos{\left(\theta_{4} \right)} \dot{\theta}_{4}^{2}\\- \ddot{\theta_2} r_{2} \cos{\left(\theta_{2} \right)} + r_{2} \sin{\left(\theta_{2} \right)} \dot{\theta}_{2}^{2} - r_{3} \sin{\left(\theta_{3} \right)} \dot{\theta}_{3}^{2} - r_{4} \sin{\left(\theta_{4} \right)} \dot{\theta}_{4}^{2}\end{matrix}\right]\right)$

````python
b = np.array( [r2*np.cos(th2)*th2p**2 - r3*np.cos(th3)*th3p**2 - r4*np.cos(th4)*th4p**2,
               r2*np.sin(th2)*th2p**2 - r3*np.sin(th3)*th3p**2 - r4*np.sin(th4)*th4p**2] )

b,A,type(th3p)
````

4. resolver con np.linalg.solve(A,b)

````python
x = np.linalg.solve(A,b)
th3pp,th4pp = x
th3pp,th4pp
````

### Verificación Numérica

````python
y2 = th4v[pos+1]
y1 = th4v[pos]
x2 = th2v[pos+1]
x1 = th2v[pos]
dth2 = x2-x1

th4pnum2 = (y2-y1)/dth2
y2 = th3v[pos+1]
y1 = th3v[pos]
th3pnum2 = (y2-y1)/dth2

y2 = th4v[pos]
y1 = th4v[pos-1]

th4pnum1 = (y2-y1)/dth2
y2 = th3v[pos]
y1 = th3v[pos-1]
th3pnum1 = (y2-y1)/dth2

ddth4_th2 = (th4pnum2-th4pnum1)/dth2
ddth3_th2 = (th3pnum2-th3pnum1)/dth2
th4ppnum = ddth4_th2*(th2p**2)
th3ppnum = ddth3_th2*(th2p**2)

th3ppnum,th4ppnum
````

### Usando Matemática Simbólica (Sympy)

````python
eq1pp = sym.diff(eq1p,t)
eq2pp = sym.diff(eq2p,t)
eq1pp,eq2pp
````

````python
eq1ppsym = eq1pp.subs([
#                 (th2sp,th2spsym),
#                 (th3sp,th3spsym),
#                 (th4sp,th4spsym),
                (th2spp,th2sppsym),
                (th3spp,th3sppsym),
                (th4spp,th4sppsym)
                   ])
eq2ppsym = eq2pp.subs([
#                 (th2sp,th2spsym),
#                 (th3sp,th3spsym),
#                 (th4sp,th4spsym),
                (th2spp,th2sppsym),
                (th3spp,th3sppsym),
                (th4spp,th4sppsym)
                    ])
eq1ppsym,eq2ppsym,th2sp,th2spsym
````

````python
eqns = [eq1ppsym,eq2ppsym]
As, bs = sym.linear_eq_to_matrix(eqns, [th3sppsym,th4sppsym])
As,bs
````

````python
values_A = {th3s(t):th3v[pos],
          th4s(t):th4v[pos],
          r3s:r3,
          r4s:r4}
values_b = {
            th2s(t):float(th2v[pos]),
            th2sp:th2p,
            th2sppsym:0,
            r2s:r2,
            th3sp:th3p,
            th4sp:th4p,
            th3s(t):th3v[pos],
            th4s(t):th4v[pos],
            r3s:r3,
            r4s:r4}
As_eval = As.evalf(subs=values_A)
bs_eval = bs.evalf(subs=values_b)
As_eval,bs_eval
````

````python
[th3pp_eval,th4pp_eval] = As_eval.LUsolve(bs_eval)
(th3pp,th3pp_eval,th3ppnum),(th4pp,th4pp_eval,th4ppnum)
````

## Análisis vectorial de aceleración



para hacer el análisis de aceleración, solo me faltan los vectores alpha de aceleración angular



````python
alpha3 = np.array([0,0,th3pp])
a3n = np.cross(w3,np.cross(w3,r3_prima_v))
a3t = np.cross(alpha3,r3_prima_v)
a2 = np.cross(w2,np.cross(w2,r2v))
a3 = a2 + a3n + a3t
````
